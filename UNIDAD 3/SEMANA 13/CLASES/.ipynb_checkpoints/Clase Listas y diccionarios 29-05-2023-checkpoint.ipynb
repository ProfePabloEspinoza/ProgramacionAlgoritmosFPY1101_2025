Ejercicio 2:
Ingresar 10 números a una lista, luego muestre la suma y promedio de los elementos.


````python
lista = []
suma = 0
promedio = 0

#Poblamos la lista
for i in range (0,10):
  numero = int(input(f'Ingrese el numero {i+1} de 10: '))
  lista.append(numero)

#Recorremos la lista, sumamos y sacamos promedio
for n in lista:
  suma = suma + n
  promedio = suma / len(lista)


print(f'La lista de numero es: {lista}')
print(f'La suma de los numeros es: {suma}')
print(f'el promedio de los numeros es: {promedio}')
````

ADIVINA QUE ANIMAL ES tienes 3 intentos


````python
import random

# Lista de animales
animales = ['perro', 'gato', 'elefante', 'león', 'tigre', 'oso', 'jirafa']

# Selecciona un animal al azar de la lista
animal_seleccionado = random.choice(animales)

# Inicia el juego
print("¡Bienvenido al juego de adivinanzas de animales!")
print("Estoy pensando en un animal. ¿Puedes adivinar cuál es? Tienes 3 intentos.")

# Contador de intentos
intentos = 0

while intentos < 3:
    # El usuario introduce su adivinanza
    adivinanza = input("Tu adivinanza: ")

    # Comprueba si la adivinanza es correcta
    if adivinanza.lower() == animal_seleccionado:
        print("¡Correcto! ¡Bien hecho!")
        break
    else:
        print("Lo siento, esa no es la respuesta correcta. ¡Inténtalo de nuevo!")
        intentos += 1

# Si el usuario no adivina el animal después de 3 intentos, revela la respuesta
if intentos == 3:
    print("Lo siento, has agotado tus intentos. El animal que estaba pensando era un " + animal_seleccionado + ".")
````

````python
alumnos = {'juanita': 68, 'pedrito': 40, 'pablito': 55}
#alumnos ['pablito']

nombre = input('Ingrese el nombre del alumno que desea saber la nota: ')
print(f'La nota de alumno  {nombre} es: {alumnos[nombre]}')
````

````python
for c,v in alumnos.items():
  print(c,v)
````

````python
# Crear un diccionario vacío para almacenar la información de los alumnos
alumnos = {}

# Función para agregar un alumno al diccionario
def agregar_alumno():
    rut = input("Ingresa el RUT del alumno: ")
    nombre = input("Ingresa el nombre del alumno: ")
    nota1 = float(input("Ingresa la primera nota del alumno: "))
    nota2 = float(input("Ingresa la segunda nota del alumno: "))
    alumnos[rut] = {"nombre": nombre, "nota1": nota1, "nota2": nota2}

# Función para buscar un alumno por su RUT
def buscar_alumno():
    rut = input("Ingresa el RUT del alumno que quieres buscar: ")
    if rut in alumnos:
        alumno = alumnos[rut]
        print(f"Alumno: {alumno['nombre']}")
        print(f"Nota 1: {alumno['nota1']}")
        print(f"Nota 2: {alumno['nota2']}")
        
    else:
        print(f"No se encontró un alumno con el RUT {rut}")

# Función para mostrar el menú y manejar la interacción del usuario
def mostrar_menu():
    while True:
        print("\n--- Menú ---")
        print("1. Agregar alumno")
        print("2. Buscar alumno")
        print("3. Salir")
        opcion = input("Elige una opción: ")

        if opcion == "1":
            agregar_alumno()
        elif opcion == "2":
            buscar_alumno()
        elif opcion == "3":
            break
        else:
            print("Opción no válida. Por favor, intenta de nuevo.")

# Mostrar el menú
mostrar_menu()
````

````python
#print(alumnos)
#print(alumnos['13483610-5'])

for c,v in alumnos.items():
  print(c,v)
````

1. `alumnos[rut]`: Aquí estamos usando el RUT del alumno como clave en el diccionario `alumnos`. En Python, los diccionarios son una estructura de datos que almacena pares de clave-valor. En este caso, estamos usando el RUT del alumno como clave porque es único para cada alumno.

2. `= {"nombre": nombre, "nota1": nota1, "nota2": nota2}`: Aquí estamos asignando un nuevo diccionario como el valor asociado a la clave RUT en el diccionario `alumnos`. Este nuevo diccionario tiene tres pares de clave-valor: "nombre" se asocia con el nombre del alumno, "nota1" se asocia con la primera nota del alumno y "nota2" se asocia con la segunda nota del alumno.

Por lo tanto, lo que esta línea de código está haciendo en su totalidad es agregar un nuevo alumno al diccionario `alumnos`, utilizando el RUT del alumno como clave y un diccionario de detalles del alumno como valor.

Por ejemplo, si llamamos a `agregar_alumno("12345678-9", "Juan Perez", 6.0, 7.0)`, el diccionario `alumnos` se verá así después de la llamada:

```python
{
    "12345678-9": {
        "nombre": "Juan Perez",
        "nota1": 6.0,
        "nota2": 7.0
    }
}
```

Y si luego llamamos a `agregar_alumno("98765432-1", "Maria Rodriguez", 5.5, 6.7)`, el diccionario `alumnos` se verá así:

```python
{
    "12345678-9": {
        "nombre": "Juan Perez",
        "nota1": 6.0,
        "nota2": 7.0
    },
    "98765432-1": {
        "nombre": "Maria Rodriguez",
        "nota1": 5.5,
        "nota2": 6.7
    }
}
```
